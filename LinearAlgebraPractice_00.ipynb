```python
import numpy as np
import sympy
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 5])
y = np.array([2, 6])
plt.plot(x, y)
plt.show()
```

Multiple points.

```python
xPoints = np.array([1, 1, 3, 3])
yPoints = np.array([0, 9, 0, 9])
plt.plot(xPoints, yPoints, color = "red")
plt.show()
```

Plot a Vector

```python
uPoint = np.array([2, 1])
v = -3*uPoint
plt.quiver(0, 0, uPoint[0], uPoint[1], angles = 'xy', scale_units = 'xy', scale = 1, color = "red")

plt.plot(0, 0, "bo")

plt.quiver(0, 0, v[0], v[1], angles = 'xy', scale_units = 'xy', scale = 1, color = "green")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
#plt.show()

# Add span of u
for c in range(-10, 10):
  span = c * uPoint
  plt.plot(span[0], span[1], "bo", 200)

plt.show()
```
